#### Задание 2

```python
t = np.linspace(start = 0, stop = 10, num = 51)
f = np.cos(t)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(t, 
        f,
        marker = 'o',
        color = 'green',
        alpha = 0.3)

ax.set(title = "График f(t)",
       xlabel = "Значения t",
       ylabel = "Значения f")

ax.set_xlim(0.5, 9.5)
ax.set_ylim(-2.5, 2.5)

plt.show()
```
